```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import pandas as pd
import string
# Import necessary libraries
import pandas as pd
from nltk.corpus import stopwords

# Create a dataframe with stopwords
stopwords_script = pd.DataFrame({'word': stopwords.words('english') + ['thats', 'weve', 'hes', 'theres', 'ive', 'im', 
                                                                       'will', 'can', 'cant', 'dont', 'youve', 'us', 
                                                                       'youre', 'youll', 'theyre', 'whats', 'didnt', 'just']})

# Function to clean the corpus
def clean_corpus(corpus):
    # Convert to lowercase
    corpus = corpus.lower()
    # Remove punctuation
    corpus = corpus.translate(str.maketrans('', '', string.punctuation))
    # Tokenize
    tokens = word_tokenize(corpus)
    # Join tokens back into cleaned text
    cleaned_text = " ".join(tokens)
    return cleaned_text

# Function to extract frequent unigrams
def frequent_terms(text):
    # Clean text data
    cleaned_text = clean_corpus(text)
    # Create a DataFrame to store word frequency
    word_freqs = pd.Series(cleaned_text.split()).value_counts()
    dm = pd.DataFrame({'word': word_freqs.index, 'freq': word_freqs.values})
    return dm

# Function to extract frequent bigrams
def frequent_bigrams(text):
    # Clean text data
    cleaned_text = clean_corpus(text)
    # Extract bigrams
    tokens = word_tokenize(cleaned_text)
    bigrams = list(ngrams(tokens, 2))
    # Create a DataFrame to store bigram frequency
    bigram_freqs = pd.Series(bigrams).value_counts()
    dm = pd.DataFrame({'word': bigram_freqs.index.map(lambda x: ' '.join(x)), 'freq': bigram_freqs.values})
    return dm

# Function to extract frequent trigrams
def frequent_trigrams(text):
    # Clean text data
    cleaned_text = clean_corpus(text)
    # Extract trigrams
    tokens = word_tokenize(cleaned_text)
    trigrams = list(ngrams(tokens, 3))
    # Create a DataFrame to store trigram frequency
    trigram_freqs = pd.Series(trigrams).value_counts()
    dm = pd.DataFrame({'word': trigram_freqs.index.map(lambda x: ' '.join(x)), 'freq': trigram_freqs.values})
    return dm
```

```python
import pandas as pd
import re
import string
from nltk import word_tokenize, ngrams
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Define the clean_text function to preprocess the text
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+", "", text)
    # Remove usernames
    text = re.sub(r"@[^\s]+", "", text)
    # Remove hashtags
    text = re.sub(r"#", "", text)
    # Remove punctuation, lowercase, and tokenize
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.lower()
    tokens = word_tokenize(text)
    # Remove stop words
    stopwords_remove = stopwords.words('english') + ['thats', 'weve', 'hes', 'theres', 'ive', 'im', 'will', 'can', 'cant', 'dont', 'youve', 'us', 'youre', 'youll', 'theyre', 'whats', 'didnt']
    tokens = [token for token in tokens if token not in stopwords_remove]
    # Join tokens back into a string
    text_cleaned = ' '.join(tokens)
    return text_cleaned

# Define the frequent_terms function to extract unigrams, bigrams, or trigrams from text
def frequent_terms(text, ngram=1, min_df=5):
    # Preprocess the text
    text_cleaned = [clean_text(t) for t in text]
    # Vectorize the text
    vectorizer = CountVectorizer(ngram_range=(ngram, ngram), min_df=min_df)
    X = vectorizer.fit_transform(text_cleaned)
    # Get the feature names and counts
    feature_names = vectorizer.get_feature_names()
    counts = X.sum(axis=0).tolist()[0]
    # Convert to a DataFrame
    df = pd.DataFrame({'term': feature_names, 'count': counts})
    df = df.sort_values(by='count', ascending=False)
    return df

# Extract unigrams from tweets
unigram = frequent_terms(tweets.loc[tweets['country'].isin(['US', 'UK', 'India', 'NoCountry'])].dropna(subset=['text'])['text'])

# Extract bigrams from tweets
bigram = frequent_terms(tweets.loc[tweets['country'].isin(['US', 'UK', 'India', 'NoCountry'])].dropna(subset=['text'])['text'], ngram=2)

# Extract trigrams from tweets
trigram = frequent_terms(tweets.loc[tweets['country'].isin(['US', 'UK', 'India', 'NoCountry'])].dropna(subset=['text'])['text'], ngram=3)

# Inspect the data
unigram.head(18)
```

```python
# split the bigram column into two words and create a new dataframe with the two words
split_bigram = bigram['term'].str.split(expand=True)
split_bigram.columns = ['word1', 'word2']

# concatenate the two dataframes and group by the two words and sum the frequencies
bigram_new = pd.concat([split_bigram, bigram['count']], axis=1)
bigram_new = bigram_new.groupby(['word1', 'word2'])['count'].sum().reset_index()

# sort by frequency in descending order
bigram_new = bigram_new.sort_values('count', ascending=False)

# display the result
bigram_new.head(18)
```

```python
# Distinct Item1
sources = bigram_new['word1'].drop_duplicates().reset_index().rename(columns={'word1': 'label'})

# Distinct item2
destinations = bigram_new['word2'].drop_duplicates().reset_index().rename(columns={'word2': 'label'})

# ----- NODES -----
# Unique Items + create unique ID
nodes = pd.concat([sources, destinations]).drop_duplicates(subset=['label']).reset_index(drop=True)
nodes['id'] = nodes.index

# ----- EDGES -----
# Adds unique ID of Item 1 to data
edges = pd.merge(bigram_new, nodes, left_on='word1', right_on='label', how='left')
edges = edges.rename(columns={'id': 'from'}).drop(columns=['label'])

# Adds unique ID of Item 2 to data
edges = pd.merge(edges, nodes, left_on='word2', right_on='label', how='left')
edges = edges.rename(columns={'id': 'to', 'count': 'weight'}).drop(columns=['label'])
edges = edges[['from', 'to', 'weight']]
nodes.head(18)
```
